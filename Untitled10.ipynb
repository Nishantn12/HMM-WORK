```python
# prompt: create a dataframe with 2 columns and 10 rows

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a DataFrame with 2 columns and 10 rows
x = np.arange(10)  # Create an array of 10 values for x
y1 = np.random.rand(10)  # Create random values for y1
y2 = np.random.rand(10)  # Create random values for y2

```

```python
plt.figure(figsize=(10,5))
plt.scatter(x,y1,c='g')
plt.scatter(x,y2,c='r')
plt.scatter(x,y2,c='b')
plt.title("Scatter plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
```

```python
plt.figure(figsize=(10,5))


plt.subplot(1,2,1)
plt.scatter(x,y1,marker=".",c="g",s=100) # Changed 'markers' to 'marker'

plt.subplot(1,2,1)
plt.scatter(x,y1,marker=".",c="r",s=100) # Changed 'markers' to 'marker'
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.title("Line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True);
plt.title("Line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True);
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y1, color='g', linestyle=':', linewidth=2)
plt.title("Line plot 1")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)


plt.subplot(1, 2, 2)
plt.plot(x, y2, color='b', linestyle='-', linewidth=2)
plt.title("Line plot 2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)


plt.show()
```

```python
names = list(student.keys())
values = list(student.values())
```
